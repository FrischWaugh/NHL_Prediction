```python
import requests
import numpy as np
import pandas as pd
from pandas.io.json import json_normalize
```

```python
url = 'https://statsapi.web.nhl.com/api/v1/statTypes'
r = requests.get(url)
json_data = r.json()
stats_type = pd.DataFrame(json_data)
stats_type
```

```python
# prenons alors
url = 'https://statsapi.web.nhl.com/api/v1/standings?season=20032004'
#url = 'https://statsapi.web.nhl.com/api/v1/teams?expand=teams.stats'
r = requests.get(url)
json_data = r.json()
print(json_data.keys())
```

```python
stats_all_raw = json_normalize(json_data["records"],record_path='teamRecords', sep = "_")
stats_all_raw.head()
```

```python
stats_all_raw.describe(include='all')
```
